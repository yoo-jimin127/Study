# 문제 1

```python
# 다음 single linked list class 가 있을 때 '결과 1'과 같이 node를 remove 하였을 때
# linked list의 size를 감소시키도록 하는 코드를 removeNode에 추가하고,
# 현재 linked list의 size를 출력하는 printSize() function을 작성하여 결과 1과 같이 출력하시오.

class Node():
    def __init__(self, data, nextNode = None):
        self.data = data
        if nextNode == None or type(nextNode) == Node :
            self.nextNode = nextNode
        else :
            print("__init__ node type error: ", type(Node))
    
    def getData(self):
        return self.data
    
    def getNextNode(self):
        return self.nextNode
    
    def setNextNode(self, node):
        if type(node) == Node:
            self.nextNode = node
        else:
            print("setNextNode node type error: " + type(Node))
            
class LinkedList:
    def __init__(self, head = None):
        self.head = head
        self.size = 0
    
    def addNode (self, data):
        newNode = Node(data, self.head)
        self.head = newNode
        self.size += 1
        return True
    
    def removeNode(self, value):
        prev = None
        curr = self.head
        while curr:
            if curr.getData() == value:
                if prev:
                    prev.setNextNode(curr.getNextNode())
                    self.size -= 1
                else:
                    self.head = curr.getNextNode()
                    self.size -= 1
                return True
            prev = curr
            curr = curr.getNextNode()
        return False
    
    def printNodes(self):
        curr = self.head
        pos = 1
        while curr:
            print("position :" + str(pos) + "  " + str(curr.data))
            pos += 1
            curr = curr.getNextNode()
            
    def printSize(self):
        print("size : " + str(self.size))

myLinkedList = LinkedList()
myLinkedList.addNode("park")
myLinkedList.addNode("choi")
myLinkedList.addNode("kim")
myLinkedList.printNodes()
myLinkedList.printSize()
print()

myLinkedList.removeNode("choi")
myLinkedList.printNodes()
myLinkedList.printSize()
```

```python
# 다음 Double Linked List는 addNode 함수에 새로운 data가 주어지면 해당 data 값을 가진 node를 Double Linked List에 추가 한다. 
# head의 유무에 따라 다음 함수가 정상적으로 작동하도록  프로그램을 완성하고, 노드를 추가하여 보이시오.

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
    def setNext(self, node):
        if type(node) == Node :
            self.next = node
            return
        elif node is None:
            self.next = None
        else :
            printf("setNext type error: {} and type is {}".format(node, type(Node)))
    
    def setPrev(self, node):
        if type(node) == Node :
            self.prev = node
            return
        elif node is None :
            self.prev = None
        else :
            print("setPrev type error: {} and type is {}".format(node, type(Node)))
            
    def getData(self) :
        return self.data
            
class DoubleLinkedList :
    def __init__(self) :
        self.head = None
        self.tail = None
        self.size = 0
    
    def getSize(self) :
        return self.size
    
    def addNode(self, data) :
        newNode = Node(data)
        if self.head is None :
            self.head = newNode
            newNode.prev = None
            newNode.next = None
            self.tail = newNode
        else :
            self.tail.setNext(newNode)
            newNode.setPrev(self.tail)
            self.tail = newNode
        self.size += 1
    
    def printNodes(self) :
        if self.size == 0 :
            print("The list is empty")
            return
        
        curr = self.head
        tail = self.tail
        #print("head = {} and tail = {}. Size = {}".format(self.head, self.tail, self.size))
        pos = 1
        
        while curr :
            print("position :" + str(pos) + "  " + str(curr.data))
            pos += 1
            curr = curr.next
   
    def printSize(self) :
        print("size : " + str(self.size))
    
myDLL = DoubleLinkedList()
print("--------before add Node--------")
myDLL.printNodes()
myDLL.printSize()
print()
print("--------after add Node---------")
myDLL.addNode("yoo")
myDLL.addNode("park")
myDLL.addNode("choi")
myDLL.addNode("kim")
myDLL.printNodes()
myDLL.printSize()
```
